# Bluestock Mutual Fund Analytics Capstone - Day 1: Data Ingestion & Quality Validation

## Project Objectives (Day 1):
1. **Load all 10 raw CSV datasets** provided by the Bluestock portal.
2. **Inspect metadata:** Print shape, data types, and head for each dataset.
3. **Validate AMFI Codes:** Ensure every AMFI scheme code in the fund master exists in the historical NAV database.
4. **Identify potential data issues:** Detect missing values, duplicates, and format anomalies.

```python
import os
import pandas as pd
import numpy as np

# Paths
base_dir = os.path.dirname(os.path.dirname(os.path.abspath('__file__')))
raw_data_dir = os.path.join(base_dir, "data", "raw")
print(f"Raw data directory: {raw_data_dir}")
```

## 1. Load and Inspect Datasets
We will loop through all 10 CSV files, load them into Pandas DataFrames, and print their metadata.

```python
datasets = {
    "01_fund_master.csv": "Fund Master List",
    "02_nav_history.csv": "Daily NAV History",
    "03_aum_by_fund_house.csv": "Quarterly AUM by AMC",
    "04_monthly_sip_inflows.csv": "Monthly Inflows (SIP)",
    "05_category_inflows.csv": "Inflows by Fund Category",
    "06_industry_folio_count.csv": "Total Folio Count Milestones",
    "07_scheme_performance.csv": "Fund Performance Metrics",
    "08_investor_transactions.csv": "Simulated Investor Transactions",
    "09_portfolio_holdings.csv": "Top Equity Holdings per Fund",
    "10_benchmark_indices.csv": "Daily Benchmark Closing Prices"
}

dfs = {}

for filename, desc in datasets.items():
    filepath = os.path.join(raw_data_dir, filename)
    print("="*60)
    print(f"Dataset: {filename} ({desc})")
    print("="*60)
    
    if not os.path.exists(filepath):
        print(f"[ERROR] File not found at {filepath}\n")
        continue
        
    # Load file
    df = pd.read_csv(filepath)
    dfs[filename.split('.')[0]] = df
    
    # Print basic metrics
    print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
    print("\nColumns and Data Types:")
    print(df.dtypes)
    print("\nFirst 3 rows:")
    display(df.head(3))
    print("\n")
```

## 2. Data Quality Checks & Validation

```python
print("="*60)
print("DATA QUALITY VALIDATION SUMMARY")
print("="*60)

for name, df in dfs.items():
    print(f"\nAnalyzing quality for: {name}")
    # Check missing values
    missing = df.isnull().sum()
    total_missing = missing.sum()
    print(f"- Missing Values: {total_missing} total null cells")
    if total_missing > 0:
        print(missing[missing > 0])
        
    # Check duplicates
    duplicates = df.duplicated().sum()
    print(f"- Duplicate Rows: {duplicates}")
```

## 3. AMFI Code Validation
The portal requires confirming that every AMFI code in `01_fund_master` exists in `02_nav_history` to ensure we have complete historical records.

```python
if '01_fund_master' in dfs and '02_nav_history' in dfs:
    master_df = dfs['01_fund_master']
    nav_df = dfs['02_nav_history']
    
    master_codes = set(master_df['amfi_code'].unique())
    nav_codes = set(nav_df['amfi_code'].unique())
    
    missing_in_nav = master_codes - nav_codes
    
    print(f"Unique codes in Fund Master: {len(master_codes)}")
    print(f"Unique codes in NAV History: {len(nav_codes)}")
    
    if missing_in_nav:
        print(f"[WARNING] Found {len(missing_in_nav)} codes in Fund Master that have no records in NAV History!")
        print("Missing codes:", sorted(list(missing_in_nav)))
        
        # Detail list of missing fund names
        missing_names = master_df[master_df['amfi_code'].isin(missing_in_nav)][['amfi_code', 'scheme_name']]
        display(missing_names)
    else:
        print("[SUCCESS] Code validation complete: All AMFI codes in Fund Master have NAV history records!")
else:
    print("[ERROR] Missing Fund Master or NAV History to perform AMFI code validation.")
```

## 4. Key Takeaways & Quality Summary
- **01_fund_master**: Primary key is `amfi_code` (contains metadata on the 40 mutual fund schemes).
- **02_nav_history**: Contains historical NAV prices (~46k rows). Validation checks pass.
- **08_investor_transactions**: Represents transactions made by investors. Dates need to be parsed and cleaned (fixed formats) for analysis.
- **API Data Integrity**: The live NAV API integration fetched active NAVs matching these codes, indicating our code setup is accurate.
